#Q6c. Design Predictor using Batch Gradient Descent Algorithm, Stochastic Gradient Algorithm and mini batch Gradient Descent algorithms with and without regularization

```python
import numpy as np
import pandas as pd
import random
import math

input_data = pd.read_csv("https://raw.githubusercontent.com/ritesh1731/ML_ASSIGNMENT/main/Housingprice")
input_data.head()
```

```python
#Use feature scaling on FloorArea
Number_of_Bedrooms = input_data['bedrooms']
Number_of_Bathrooms = input_data['bathrms']
Price = input_data['price']
FloorArea = input_data['lotsize']

#Feature scaling

Mean_of_FloorArea = np.mean(FloorArea)
Max_of_FloorArea = max(FloorArea)
Min_of_FloorArea = min(FloorArea)
FloorArea_Scaled = []
for i in FloorArea:
	FloorArea_Scaled.append((i - Mean_of_FloorArea) / (Max_of_FloorArea - Min_of_FloorArea))
 

 #Setting up training and test data
FloorArea_Train = FloorArea_Scaled[:383]

Number_of_BathroomsTrain = Number_of_Bathrooms[:383]
Number_of_BedroomsTrain = Number_of_Bedrooms[:383]
PriceTrain = Price[:383]
FloorAreaTest = FloorArea[383:]
Number_of_BathroomsTest = Number_of_Bathrooms[383:]
Number_of_BedroomsTest = Number_of_Bedrooms[383:]
Price_Test = Price[383:]

```

```python
#Batch gradient descent without regularization 
X_Train = []
for i in range(383):
	X_Train.append([1, FloorArea_Scaled[i], Number_of_Bedrooms[i], Number_of_Bathrooms[i]])
 
X_Test = []
n = len(FloorAreaTest)
for i in range(383,383+n):
  X_Test.append([1,FloorArea_Scaled[i],Number_of_Bedrooms[i],Number_of_Bathrooms[i]])


x = np.array(X_Train)
y = np.array(PriceTrain)
w = np.array(([0],[0],[0],[0]))
y = y.reshape(len(y),1)
# print(w)

def product(w,x):
    ans=0
    for i in range(w.shape[0]):
        ans+=w[i][0]*x[i]
    return ans

def gradient(x,w,y):
    ans=np.zeros(w.shape[0])
    for i in range(x.shape[0]):
        ans=(ans+(product(w,x[i])-y[i][0])*x[i])
    ans=ans.reshape(w.shape[0],1)
    return ans

lr=0.001
for i in range(5000):
  w = w-lr*gradient(x,w,y)/x.shape[0]

print("Cefficients are ",w)

# Mean absolute percentage error
x_test = np.array(X_Test)
y_test = np.array(Price_Test)
y_pred = np.dot(x_test,w)

n = x_test.shape[0]
mat = 0

for i in range(n):
  mat = mat + abs(y_pred[i][0]-y_test[i])/y_test[i]
mat = mat*100/n
print("Mean absolute percentage error= ",mat)
```

```python
#Batch gradient descent with regularization 

X_Train= []
for i in range(383):
	X_Train.append([1, FloorArea_Scaled[i], Number_of_Bedrooms[i], Number_of_Bathrooms[i]])
 
X_Test = []
n = len(FloorAreaTest)
for i in range(383,383+n):
  X_Test.append([1,FloorArea_Scaled[i],Number_of_Bedrooms[i],Number_of_Bathrooms[i]])

x = np.array(X_Train)
y = np.array(PriceTrain)
w = np.array(([0],[0],[0],[0]))
y = y.reshape(len(y),1)
# print(w)

def product(w,x):
    ans=0
    for i in range(w.shape[0]):
        ans+=w[i][0]*x[i]
    return ans

def gradient(x,w,y):
    ans=np.zeros(w.shape[0])
    for i in range(x.shape[0]):
        ans=(ans+(product(w,x[i])-y[i][0])*x[i])
    ans=ans.reshape(w.shape[0],1)
    return ans

lamda = -120
lr=0.001
m = x.shape[0]
for i in range(5000):
  temp = 1-((lr*lamda)/m)
  w = temp*w-lr*gradient(x,w,y)/m

print("Cefficients are ",w)


# mean absolute percentage error
x_test = np.array(X_Test)
y_test = np.array(Price_Test)
y_pred = np.dot(x_test,w)

n = x_test.shape[0]
mat = 0

for i in range(n):
  mat = mat + abs(y_pred[i][0]-y_test[i])/y_test[i]
mat = mat*100/n
print("Mean absolute percentage error= ",mat)
```

```python
#Stochastic gradient descent without regularization 
X_Train= []
for i in range(383):
	X_Train.append([1, FloorArea_Scaled[i], Number_of_Bedrooms[i], Number_of_Bathrooms[i]])
 
X_Test = []
n = len(FloorAreaTest)
for i in range(383,383+n):
  X_Test.append([1,FloorArea_Scaled[i],Number_of_Bedrooms[i],Number_of_Bathrooms[i]])

x = np.array(X_Train)
y = np.array(PriceTrain)
w = np.array([0,0,0,0])
y = y.reshape(len(y),1)

print("initial coefficients : ",w)

def product(w,x):
    ans=0
    for i in range(x.shape[0]):
        ans= ans +w[i]*x[i]
    return ans


lr=0.0005
for itr in range(10):
  for i in range(x.shape[0]):
    grad = (product(w,x[i]) - y[i][0])*x[i]
    w = w - lr*grad

print("coefficients are :",w)


# mean absolute percentage error
w = w.reshape((4,1))
x_test = np.array(X_Test)
y_test = np.array(Price_Test)
y_pred = np.dot(x_test,w)

n = x_test.shape[0]
mat = 0

for i in range(n):
  mat = mat + abs(y_pred[i][0]-y_test[i])/y_test[i]
mat = mat*100/n
print("mean absolute percentage error",mat)
```

```python
#Stochastic gradient descent with regulariation
X_Train= []
for i in range(383):
	X_Train.append([1, FloorArea_Scaled[i], Number_of_Bedrooms[i], Number_of_Bathrooms[i]])
 
X_Test = []
n = len(FloorAreaTest)
for i in range(383,383+n):
  X_Test.append([1,FloorArea_Scaled[i],Number_of_Bedrooms[i],Number_of_Bathrooms[i]])

x = np.array(X_Train)
y = np.array(PriceTrain)
w = np.array([0,0,0,0])
y = y.reshape(len(y),1)

print("initial coefficients : ",w)

def product(w,x):
    ans=0
    for i in range(x.shape[0]):
        ans= ans +w[i]*x[i]
    return ans


lr=0.0005
lamda = -30
for itr in range(10):
  for i in range(x.shape[0]):
    grad = (product(w,x[i]) - y[i][0])*x[i]
    temp = 1-((lamda*lr)/x.shape[0])
    w = temp*w - lr*grad

print("coefficients are :",w)


# mean absolute percentage error
w = w.reshape((4,1))
x_test = np.array(X_Test)
y_test = np.array(Price_Test)
y_pred = np.dot(x_test,w)

n = x_test.shape[0]
mat = 0

for i in range(n):
  mat = mat + abs(y_pred[i][0]-y_test[i])/y_test[i]
mat = mat*100/n
print("Mean absolute percentage error",mat)
```

```python
#Mini batch gradient descent without regularization 
X_Train= []
for i in range(383):
	X_Train.append([1, FloorArea_Scaled[i], Number_of_Bedrooms[i], Number_of_Bathrooms[i]])
 
X_Test = []
n = len(FloorAreaTest)
for i in range(383,383+n):
  X_Test.append([1,FloorArea_Scaled[i],Number_of_Bedrooms[i],Number_of_Bathrooms[i]])

bs = 20;
lr = 0.002

x = np.array(X_Train)
y = np.array(PriceTrain)
w = np.array([0,0,0,0])
y = y.reshape(len(y),1)

def product(w,x):
    ans=0
    for i in range(x.shape[0]):
        ans= ans +w[i]*x[i]
    return ans

def gradient(x,w,y,st,bs):
    ans=np.zeros(w.shape[0])
    for i in range(st,st+bs):
        if(i>=x.shape[0]):
          break
        ans=(ans+(product(w,x[i])-y[i][0])*x[i])
    return ans

for itr in range(50):
  for i in range(bs):
    w = w - (lr*gradient(x,w,y,bs*i,bs))/bs
print("coefficients are: ",w)



# mean absolute percentage error
w = w.reshape((4,1))
x_test = np.array(X_Test)
y_test = np.array(Price_Test)
y_pred = np.dot(x_test,w)

n = x_test.shape[0]
mat = 0

for i in range(n):
  mat = mat + abs(y_pred[i][0]-y_test[i])/y_test[i]
mat = mat*100/n
print("Mean absolute percentage error=",mat)
```
